```python
import numpy as np
import matplotlib.pyplot as plt

N = 64
x = np.arange(N)
mu = N // 2
sigma = 5

gaussian = np.exp(-0.5 * ((x - mu) / sigma) ** 2)

sine_wave_small = 0.02 * np.sin(2 * np.pi * x / N * 2)
gaussian_with_small_sine = gaussian + sine_wave_small

noise = np.random.normal(0, 0.02, N)
gaussian_with_sine_and_noise = gaussian_with_small_sine + noise

fft_final = np.fft.fft(gaussian_with_sine_and_noise)
ifft_final = np.fft.ifft(fft_final).real
ifft_final_shifted = np.fft.fftshift(ifft_final)

fig, axs = plt.subplots(3, 1, figsize=(10, 6), sharex=True)

axs[0].plot(gaussian_with_sine_and_noise)
axs[0].set_title('Original: Gaussian + 2% Sine + 2% Noise')

axs[1].plot(ifft_final)
axs[1].set_title('IFFT Result (Offset present)')

axs[2].plot(ifft_final_shifted)
axs[2].set_title('IFFT Result with fftshift (Peak centered)')

plt.tight_layout()
plt.savefig('../Plot/IFFT Result with fftshift.pdf', dpi = 300)
plt.show()
```

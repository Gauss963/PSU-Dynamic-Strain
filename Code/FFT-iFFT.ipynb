```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 60
```

```python
x = np.arange(N)
mu = N // 2
sigma = 5
gaussian = np.exp(-0.5 * ((x - mu) / sigma) ** 2)

fft_gaussian = np.fft.fft(gaussian)
ifft_gaussian = np.fft.ifft(fft_gaussian).real
ifft_gaussian_shifted = np.fft.fftshift(ifft_gaussian)

fig, axs = plt.subplots(3, 1, figsize=(10, 6), sharex=True)

axs[0].plot(gaussian)
axs[0].set_title('Original Gaussian (Peak in the middle)')

axs[1].plot(ifft_gaussian)
axs[1].set_title('IFFT Result (Peak shifted to front)')

axs[2].plot(ifft_gaussian_shifted)
axs[2].set_title('IFFT Result with fftshift (Peak back to middle)')

plt.tight_layout()
plt.savefig('../Plot/IFFT Result with fftshift.pdf', dpi = 300)
plt.show()
```
